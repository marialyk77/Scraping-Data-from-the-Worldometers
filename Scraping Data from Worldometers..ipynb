```python
from bs4 import BeautifulSoup
import requests
```

```python
# The requests.get(url) returns a response object with all infos about the request. 

# The result of the request.get(url) is assigned to the variable 'page' to facilitate the work. 

# The variable 'soup' is assigned the BeautifulSoup object, created from the HTML content of the web page stored in the 'page' variable. 
# This BeautifulSoup object provides a convenient way to navigate and extract specific information from the HTML structure using methods like soup.find().
```

```python
url = 'https://www.worldometers.info/world-population/population-by-country/'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
# better to pull the table that we need and SAVE it is as 'table'. 
```

```python
table = soup.find('table', 'table table-striped table-bordered')
```

```python
table
```

```python
# pulling the columns 
```

```python
table.find_all('th')
```

```python
# Because we need the info with the columns names, better to put it in a list and get rif od the breakes (br) and tags (th).
```

```python
world_columns = table.find_all('th')
```

```python
world_column_titles = [ columns.text for columns in world_columns]

print(world_column_titles)
```

```python
# before moving on we better put the column names in a df 
```

```python
import pandas as pd

df = pd.DataFrame(columns = world_column_titles)

df
```

```python
# Pulling the info from rows, based on the current dataon 01/02/2024, India was ranked first! And we see that below too! 
```

```python
table. find_all('td')
```

```python
table. find_all('tr')
```

```python
# we use tr because we can search for td tage within the tr tag. The tr tag refers to the actual row. Each row (tr) inside of it, it has the td tags. 
```

```python
world_rows = table. find_all('tr')
```

```python
for row in world_rows:
    row_data = row.find_all('td')
    individual_row_data = [data.text for data in row_data]
    
    print(individual_row_data)
    
```

```python
# Next is to save the row data to a pandas df! So,1) we are going to pull in the data at each level, 2) we are going to append each row to the data frame and then loop back through for the next row. And we should start at the 1st row, not the 0 row - because that is the column.
```

```python
# 1. We are going to iterate though each row except from the 1st one, which is for the column. 
# 2. We are trying to find all the td tags. 
# 3. Appending to the list: individual_row_data. 
# 4. Each time we are passing through we are looking at the length of the df , and ask to place the individual_row_data at this index position: df.loc(length)
# 5. printing the df 
```

```python
for row in world_rows[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text for data in row_data]
     
    lenght = len(df)
    df.loc[lenght] = individual_row_data
    
    print(df)
    
    
```
